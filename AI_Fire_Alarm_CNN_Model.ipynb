```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8n.pt")
```

```python
results = model.train(
    data="/content/Fire_Dataset/data.yaml",
    epochs=100,
    imgsz=640,
    batch=16,
    project="Fire_Project",
    name="Fire_Detection",
    device=0
)
```
